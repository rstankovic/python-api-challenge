```python
import requests
import gmaps
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
import json
from config import weatherKey
from citipy import citipy
import pandas as pd
from config import googleKey
import random


#just to have the JSON pulled up always
# city = citipy.nearest_city(-85, 150)
# print(city.city_name, city.country_code)
# baseURL = "http://api.openweathermap.org/data/2.5/weather?"
# extendURL = f"q={city.city_name},{city.country_code}&appid={weatherKey}"

# responseJSON = requests.get(baseURL + extendURL).json()
# print(json.dumps(responseJSON,indent = 4))
```

```python
'''
we will search through latitudes to find nearest cities and their weather over the past month (which is all we get
from the openweather api).
we will loop through latitudes, checking through the entire latitude (via longitude).
will perform stats calulations such as mean, median, highs, lows, etc
will also map these temperatures to a map using gmaps
'''
#first get base URL
baseURL = "http://api.openweathermap.org/data/2.5/weather?"
'''
then loop through lat and long to find cities. as the cities come in, we make requests to
open weather api (500 calls is unavoidable, each city must be investigated)
before moving on, we analyze the data since storing that data and analyzing it later will result in a loooot
of data that we would only use once and would not need to refer back to multiple times.
would obviously store important data but only averages and the like- not every single data point
'''
#latidues: -90 to 90
#longitude: 0 to 180
#build the dictionary
city_dict = {}
latStats = {}
count = 0
latRange = np.arange(-90,90,5)
lngRange = np.arange(-180,180,10)
city_uniques = []
for lat in latRange:
    #for statistical analysis
    #print('lat')
    tempsK = []
    latHumidity = []
    latCloud = []
    latWind = []
    latStats[f"{lat}"] = {}
    for lng in lngRange:
        #print("lng")
        city = citipy.nearest_city(lat, lng)
        if city.city_name in city_uniques:
            continue
        else:
            city_uniques.append(city.city_name)
            extendURL = f"q={city.city_name},{city.country_code}&appid={weatherKey}"
            responseJSON = requests.get(baseURL + extendURL).json()
            try:
                city_dict[city.city_name] = {}
                city_dict[city.city_name]["tempK"] = responseJSON["main"]["temp"]
                city_dict[city.city_name]["humidity"] = responseJSON["main"]["humidity"]
                city_dict[city.city_name]["wind speed"] = responseJSON["wind"]["speed"]
                city_dict[city.city_name]["clouds"] = responseJSON["clouds"]["all"]
                tempsK.append(responseJSON["main"]["temp"])
                latHumidity.append(responseJSON["main"]["humidity"])
                latCloud.append(responseJSON["clouds"]["all"])
                latWind.append(responseJSON["wind"]["speed"])
                city_dict[city.city_name]["lat"] = responseJSON["coord"]["lat"]
                city_dict[city.city_name]["lon"] = responseJSON["coord"]["lon"]
                print(city.city_name)
                count+=1
                
            except KeyError:
                #print(f"{city.city_name} not found")
                continue
        
            
        #print(json.dumps(responseJSON,indent = 4))
        
    latStats[f"{lat}"]["Average Temp"] = np.average(tempsK)
    latStats[f"{lat}"]["Average Humidity"] = np.average(latHumidity)
    latStats[f"{lat}"]["Average Cloudiness"] = np.average(latCloud)
    latStats[f"{lat}"]["Average Wind Speed (mph)"] = np.average(latWind)
print(len(city_uniques))
```

```python
city_df = pd.DataFrame(city_dict).T
city_df = city_df.dropna()
city_df["lat"] = city_df["lat"].astype(float)
city_df.head()
```

```python
plt.scatter(x=lat_df.index,y=lat_df["Average Temp"])
plt.xlabel("latitude")
plt.ylabel("Average Temp")
plt.title("latitude vs Average Temp")
plt.figure(figsize = (10,3))
plt.show()
print("Average temperature spikes a lot when near the equator. keep in mind that the \
scewering towards the right-hand side\n is because fewer cities lay near the northpole than the south pole")
plt.scatter(x=lat_df.index,y=lat_df["Average Humidity"])
plt.xlabel("latitude")
plt.ylabel("Average Humidity (%)")
plt.title("latitude vs Average Humidity (%)")
plt.show()
print("significant drop in humidity around the equator")
plt.scatter(x=lat_df.index,y=lat_df["Average Cloudiness"])
plt.xlabel("latitude")
plt.ylabel("Cloudiness (%)")
plt.title("latitude vs Cloudiness (%)")
plt.show()
print("clouds apparently cease to exist in the southern-most part of the globe. although this is all\
near africa (map is provided at the bottom of all the chosen cities)")
plt.scatter(x=lat_df.index,y=lat_df["Average Wind Speed (mph)"])
plt.xlabel("latitude")
plt.ylabel("Average Wind Speed (mph)")
plt.title("latitude vs Average Wind Speed (mph)")
plt.show()
print("wind speed tends to scatter around the equator and in the north. check the map for consistency")
```

```python
lat_df["lat"] = lat_df.index
lat_df["lat"] = pd.to_numeric(lat_df["lat"])

for statistic, column_data in lat_df.iteritems():
    if statistic == "lat":
        break
    #northern hemisphere
    xvalues = lat_df.loc[lat_df["lat"] > 0,"lat"]
    yvalues = lat_df.loc[lat_df["lat"] > 0,statistic]
    #print(xvalues)
    (slope, intercept, rvalue, pvalue, stderr) = sts.linregress(xvalues, yvalues)
    regress_values = xvalues * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(xvalues,yvalues)
    plt.plot(xvalues,regress_values,"r-")
    plt.annotate(line_eq,(6,10),fontsize=15,color="red")
    plt.xlabel('Lat (northern hemisphere)')
    plt.ylabel(statistic)
    plt.show()
    #souther hemisphere
    xvalues = lat_df.loc[lat_df["lat"] < 0,"lat"]
    yvalues = lat_df.loc[lat_df["lat"] < 0,statistic]
#     print(xvalues)
    (slope, intercept, rvalue, pvalue, stderr) = sts.linregress(xvalues, yvalues)
    regress_values = xvalues * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(xvalues,yvalues)
    plt.plot(xvalues,regress_values,"r-")
    plt.annotate(line_eq,(6,10),fontsize=15,color="red")
    plt.xlabel('Lat (southern hemisphere)')
    plt.ylabel(statistic)
    plt.show()
```

```python
coords = []
for index,row in city_df.iterrows():
    coords.append(tuple([row.lat,row.lon]))
figure_layout = {
    'width': '1000px',
    'height': '1000px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)

heat_layer = gmaps.heatmap_layer(coords, weights=city_df["tempK"]/5)
heat_layer.max_intensity = 100
heat_layer.point_radius = 5
fig.add_layer(heat_layer)
city_layer = gmaps.symbol_layer(
    coords, fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=2,
    info_box_content=[f"Temp of city: {cityTemp}" for cityTemp in city_df["tempK"]]
)
fig.add_layer(city_layer)
fig
```
